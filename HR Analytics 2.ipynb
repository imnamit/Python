```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('IBM_Attrition_Dataset.csv')
data
```

```python
print (data)
```

```python
attrition = data.groupby('Attrition')
attrition.mean()
```

```python
from sklearn.cluster import KMeans
import pandas as pd
```

```python
#Filter data
attrition_emp=data[['MonthlyIncome','HourlyRate']][data.Attrition == 'Yes']
```

```python
#Create group using K-means clustering
kmeans=KMeans(n_clusters=3,random_state=0).fit(attrition_emp)
kmeans
```

```python
#Add new column "label" and assign cluster labels.
attrition_emp["label"]=kmeans.labels_
```

```python
#Draw scatter plot
plt.scatter(attrition_emp["MonthlyIncome"],attrition_emp["HourlyRate"], c=attrition_emp["label"], cmap="Accent")
plt.xlabel("MonthlyIncome")
plt.ylabel("HourlyRate")
plt.title("3 Clusters of employee who left")
plt.show()
```
